```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## Local Government Area Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average maths score 

* Calculate the average reading score

* Calculate the percentage of students with a passing maths score (50 or greater)

* Calculate the percentage of students with a passing reading score (50 or greater)

* Calculate the percentage of students who passed maths **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Count the total number of schools
total_schools = school_data["school_name"].count()
# Count the total number of students
total_students = student_data["student_name"].count()
# Calculate total budget
total_budget = school_data["budget"].sum()
# Average maths score
ave_maths_score = student_data["maths_score"].mean()
# Average reading score
ave_read_score = student_data["reading_score"].mean()
# Percentage of students with a passing maths score (50 or greater)
passing_maths = ((school_data_complete["maths_score"] >= 50).sum() / total_students) * 100
# Percentage of students with a passing reading score (50 or greater)
passing_read = ((school_data_complete["reading_score"] >= 50).sum() / total_students) * 100
# Percentage of students who passed maths and reading (% Overall Passing)
passing_df = school_data_complete.loc[(school_data_complete["maths_score"] >= 50) & 
                                      (school_data_complete["reading_score"] >= 50)]
passing_overall = (passing_df["student_name"].count() / total_students) * 100

# Create a dataframe
local_area_summary = pd.DataFrame(
    {"Total Schools": [total_schools],
     "Total Students": [total_students],
     "Total Budget": [total_budget],
     "Average Maths Score": [ave_maths_score],
     "Average Reading Score": [ave_read_score],
     "% Passing Maths": [passing_maths],
     "% Passing Reading": [passing_read],
     "% Overall Passing": [passing_overall]
     }
)
```

```python
# Format the data 
local_area_summary["Total Students"] = local_area_summary["Total Students"].map("{:,}".format)
local_area_summary["Total Budget"] = local_area_summary["Total Budget"].map("${:,}".format)
local_area_summary
```

## School Summary

* Create an overview table that summarises key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed maths **and** reading.)
  
* Create a dataframe to hold the above results

```python
# Create a new data frame with some of the previous data frame's columns
school_summary = school_data[["school_name","type","size","budget"]]
# Rename columns
school_summary = school_summary.rename(columns={"type":"School Type", "size":"Total Students", "budget":"Total School Budget"})
# Per Student Budget
school_summary["Per Student Budget"] = school_summary["Total School Budget"] / school_summary["Total Students"]
# Organise data frames for each school 
Huang = (student_data.loc[student_data["school_name"] == "Huang High School"])
Figueroa = (student_data.loc[student_data["school_name"] == "Figueroa High School"])
Shelton = (student_data.loc[student_data["school_name"] == "Shelton High School"])
Hernandez = (student_data.loc[student_data["school_name"] == "Hernandez High School"])
Griffin = (student_data.loc[student_data["school_name"] == "Griffin High School"])
Wilson = (student_data.loc[student_data["school_name"] == "Wilson High School"])
Cabrera = (student_data.loc[student_data["school_name"] == "Cabrera High School"])
Bailey = (student_data.loc[student_data["school_name"] == "Bailey High School"])
Holden = (student_data.loc[student_data["school_name"] == "Holden High School"])
Pena = (student_data.loc[student_data["school_name"] == "Pena High School"])
Wright = (student_data.loc[student_data["school_name"] == "Wright High School"])
Rodriguez = (student_data.loc[student_data["school_name"] == "Rodriguez High School"])
Johnson = (student_data.loc[student_data["school_name"] == "Johnson High School"])
Ford = (student_data.loc[student_data["school_name"] == "Ford High School"])
Thomas = (student_data.loc[student_data["school_name"] == "Thomas High School"])

school_list = [Bailey,Cabrera,Figueroa,Ford,Griffin,Hernandez,Holden,Huang,Johnson,Pena,Rodriguez,Shelton,Thomas,Wilson,Wright]

# Arrange in alphabetical order and set the school names as the row indices
school_summary.sort_values('school_name',ascending=True,inplace=True)
school_summary.set_index('school_name',inplace=True)

# Average maths score
average_maths = [school["maths_score"].mean() for school in school_list]
school_summary["Average Maths Score"] = average_maths

# Average reading score
average_reading = [school["reading_score"].mean() for school in school_list]
school_summary["Average Reading Score"] = average_reading

# % Passing Maths
maths_pass = []
for school in school_list:
    maths_pass.append(((school["maths_score"] >= 50).sum() / school["student_name"].count())*100)
school_summary["% Passing Maths"] = maths_pass

# % Passing Reading
reading_pass = []
for school in school_list:
    reading_pass.append(((school["reading_score"] >= 50).sum() / school["student_name"].count())*100)

school_summary["% Passing Reading"] = reading_pass
# % Overall Passing

overall_pass = []
for school in school_list:
    passing_df2= school.loc[(school["maths_score"] >= 50) & 
                                      (school["reading_score"] >= 50)]
    overall_pass.append((passing_df2["student_name"].count() / school["student_name"].count()) * 100)

school_summary["% Overall Passing"] = overall_pass

# Create datafiles for a future question that requires the budget as an integer
spend_df = school_summary.loc[:,("Per Student Budget", "Average Maths Score", "Average Reading Score", 
                          "% Passing Maths", "% Passing Reading", "% Overall Passing")]
size_df = school_summary.loc[:,("Total Students", "Average Maths Score", "Average Reading Score",
                                "% Passing Maths", "% Passing Reading", "% Overall Passing")]

# Format budget columns
school_summary["Total School Budget"] = school_summary["Total School Budget"].map("${:,.2f}".format)
school_summary["Per Student Budget"] = school_summary["Per Student Budget"].map("${:,.2f}".format)
    
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Create bins and labels
spend_bins = [0, 585, 630, 645, 680]
spend_labels = ["<$585", "$585-630", "$630-645", "$645-680"]
# Add an extra column with the bins and organise by average
spend_df["Spending Ranges (Per Student)"] = pd.cut(spend_df["Per Student Budget"], spend_bins, labels=spend_labels)
grouped_spend_df = spend_df.groupby(["Spending Ranges (Per Student)"])
grouped_spend_df = grouped_spend_df.mean()
# Delete "Per Student Budget" column
del grouped_spend_df["Per Student Budget"]
# Format columns to be two decimal places
grouped_spend_df["Average Maths Score"] = grouped_spend_df["Average Maths Score"].map("{:.2f}".format)
grouped_spend_df["Average Reading Score"] = grouped_spend_df["Average Reading Score"].map("{:.2f}".format)
grouped_spend_df["% Passing Maths"] = grouped_spend_df["% Passing Maths"].map("{:.2f}".format)
grouped_spend_df["% Passing Reading"] = grouped_spend_df["% Passing Reading"].map("{:.2f}".format)
grouped_spend_df["% Overall Passing"] = grouped_spend_df["% Overall Passing"].map("{:.2f}".format)
grouped_spend_df
```
